## Task 1
Convert gray-scale image to binary using a threshold

```python
binary_image = np.zeros((gray_image.shape[0], gray_image.shape[1]), dtype='float64')

def img_to_bin(gray_img):
  # start of your code here
     # img=gray_img
      img= get_size(gray_img)
      row = img[0]
      col = img[1]
      array_2d = [[0 for j in range(col)] for i in range(row)]
      pixel_array= array_2d
      for i in range(row):
       for j in range(col):
         pixel_value = get_pixel_value(gray_img, i, j)
         if pixel_value < 128:
          pixel_array[i][j] = 0
         else:
          pixel_array[i][j]=1


       return pixel_array
  # end of your code here
```
